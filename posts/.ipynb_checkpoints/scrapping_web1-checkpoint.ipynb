-Importar los paquetes necesarios.

```python
import requests
from bs4 import BeautifulSoup
```

Si no conoces estos paquetes puedes preguntar al Señor Google por ellos. "requests, BeautifulSoup".

-Especificar la url a peinar.

```python
url = 'http://www.uca.edu.ni/index.php/carreras'
```

-Generar una variable llamada r que capture la validación de la url. Si es 200 indica que la página existe.

```python
r = requests.get(url)
print(r)
```

-Extraer la respuesta como html

```python
html_doc = r.text
```

-Embellecer con BeautifulSoup el html extraído.

```python
soup = BeautifulSoup(html_doc, 'lxml')
```

-Verificar si el título de la página es correcto

```python
print(soup.title)
```

-Echemos un vistazo a lo que deberíamos extraer:

```python
print(soup)
```

-Obteniendo el contenido como textos en una lista

```python
lista = []
for ul in soup.find_all('a'):
    texto = ul.get_text('href')
    lista.append(texto)
```

-En esta parte filtramos la información que queremos

```python
carreras = lista[54:73]
carreras
```

-Veamos algunos métodos y funciones con la lista:

--Ordenando notamos que tienen Arquitectura, cinco ingenierías y trece licenciaturas

```python
carreras.sort()
carreras
```

--El número de carreras que posee la UCA son 19.

```python
len(carreras)
```

--Tiene 13 licenciaturas

```python
import re
licenciaturas = re.findall('Licenciatura', str(carreras))
len(licenciaturas)
```

--Tiene 5 Ingenierías

```python
ingenierias = re.findall('Ingeniería', str(carreras))
len(ingenierias)
```

-- 1 es Arquitectura

--Observar la lista

```python
for i in carreras:
    print(i.capitalize())
```

--Grafiquemos en un gráfico de barras

```python
titulacion = ['Licenciatura', 'Ingeniería', 'Arquitectura']
cantidad = [len(licenciaturas), len(ingenierias), 1]
tabla = {'Titulacion':titulacion, 'Cantidad':cantidad}
import pandas as pd
tabla = pd.DataFrame(tabla)
tabla
```

```python
import matplotlib.pyplot as plt
tabla.plot.bar(x='Titulacion', y='Cantidad', rot=0)
```
